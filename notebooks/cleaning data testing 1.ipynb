# Cleaning of the data

```python
import pandas as pd
import matplotlib.pyplot as plt
salary_df = pd.read_csv('Salary_Data.csv') #reading the csv file
salary_df.dropna(inplace = True) # remove all null rows
salary_df = salary_df.sort_values('Age') # arranges or sorts in order
salary_df.drop_duplicates(inplace = True) # removes all duplicate rows
salary_df['Gender'].str.strip("123._/") # strips the Gender column and numerical or special keys
salary_df['Education Level'].str.strip("123._/")
salary_df['Job Title'].str.strip("123._/")
salary_df["Education Level"] = salary_df["Education Level"].str.replace("Degree","") # replace the degree with an empty space
salary_df["Education Level"] = salary_df["Education Level"].str.strip() # strips the empty space 
salary_df["Education Level"] = salary_df["Education Level"].str.replace("Master's","Master's Degree ") # replace the Master's with Master's Degree
salary_df["Education Level"] = salary_df["Education Level"].str.replace("phD","PhD") # replace the phD with PhD
salary_df["Education Level"] = salary_df["Education Level"].str.replace("Bachelor's","Bachelor's Degree ") # replace the dBachelor's with an Bachelor's Degree
# salary_df.to_csv("Clean Database.csv") Save the changes to a new file

salary_df
```

# Modelling of data

```python
import pandas as pd

# melbourne_file_path = '../input/melbourne-housing-snapshot/melb_data.csv'
clean_data = pd.read_csv('Clean Database.csv') 
clean_data.columns
```

```python
y = clean_data.Salary
```

```python
clean_data_features = ['Age','Years of Experience']
```

```python
X = clean_data[clean_data_features]
```

```python
X.describe()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
salary_model = DecisionTreeRegressor(random_state=1)

# Fit model
salary_model.fit(X, y)
```

```python
print("Making predictions for the salary:")
print(X.head())
print("The predictions are")
print(salary_model.predict(X.head()))
```

# Model Validation

### Once we have a model, we calculate the mean absolute error

```python
#  Mean Absolute Error

from sklearn.metrics import mean_absolute_error

predicted_Salary = salary_model.predict(X)
mean_absolute_error(y, predicted_Salary)
```

```python
from sklearn.model_selection import train_test_split # split data into training and validation data, for both features and target

# The split is based on a random number generator. Supplying a numeric value to the random_state argument guarantees we get the same split every time we run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
salary_model = DecisionTreeRegressor() # Define model
salary_model.fit(train_X, train_y)# Fit model

# get predicted prices on validation data
validation_predictions = salary_model.predict(val_X)
print(mean_absolute_error(val_y, validation_predictions))
```

### where a model matches the training data almost perfectly, but does poorly in validation and other new data is called overfitting while underfitting is When a model fails to capture important distinctions and patterns in the data, so it performs poorly even in training data

```python
# from sklearn.metrics import mean_absolute_error
# from sklearn.tree import DecisionTreeRegressor

def get_mean_abso_error(max_leaf_nodes, train_X, val_X, train_y, val_y):
    salary_model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    salary_model.fit(train_X, train_y)
    predicts_validation = salary_model.predict(val_X)
    mean_abso_error = mean_absolute_error(val_y, predicts_validation)
    return(mean_abso_error)
#  The data is loaded into train_X, val_X, train_y and val_y 
```
